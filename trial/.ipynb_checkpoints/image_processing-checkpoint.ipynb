### スペクトラム分析
- スペクトラム分析，いわゆる周波数分析もできる
- 通常のスペクトラム分析で使われるのは離散フーリエ変換．
- しかし，離散フーリエ変換は時間の概念がないため，インパルス信号やノイズに弱い傾向にある．
- そのため、短時間で区切りフーリエ変換をかける短時間フーリエ変換（STFT）を行うことが多い．

```python
D= librosa.stft(music)
print(D)
```

```python
X = librosa.stft(music)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=fs,x_axis='time',y_axis='hz')
plt.colorbar()
```

### ケプストラム

```python
from librosa.display import specshow
```

```python
mfccs = librosa.feature.mfcc(music[fs:fs*2], sr=fs)


specshow(mfccs, sr=fs, x_axis='time')
plt.colorbar()
```

- これが1秒から2秒のケプストラム

- 音楽データを切り取る関数

```python
def cut_music(data, fs, cut_time=1):
    
    cut_time = cut_time
    
    cut_data = []
    for t in range(len(data) // (cut_time*fs)):
        cut_data.append(data[fs*cut_time*t:fs*cut_time*(t+1)])
        
    cut_data.append(data[len(data) // (cut_time*fs) - 1:])
    
    return cut_data
```

```python
len(cut_music(music))
```
